```python
import random as r

original_set = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}
original_list = list(original_set)  # Convert set to list for random.sample

SetSize = 5
Iterations = 10000
ResultSet = set()

for _ in range(Iterations):
    subset = r.sample(original_list, SetSize)  # Use list here
    if sum(subset) == 0:
        ResultSet.add(tuple(sorted(subset)))

for res in ResultSet:
    print(res)

print("\nTotal subsets found:", len(ResultSet))
```

```python
import random as r

original_set = {-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4}
original_list = list(original_set)

min_size = 3
max_size = 6
iterations = 20000
result_set = set()

for _ in range(iterations):
    size = r.randint(min_size, max_size)  # Random subset size between 3 and 6
    subset = r.sample(original_list, size)  # Randomly pick 'size' elements
    if sum(subset) == 0:
        result_set.add(tuple(sorted(subset)))  # Add sorted tuple to avoid duplicates

# Print results
for res in sorted(result_set):
    print(res)

print("\nTotal subsets found:", len(result_set))
```
